```python
from sklearn.preprocessing import MinMaxScaler

def normalize_numerical_columns(df_train, df_test):
  df_train_normalized = df_train.copy()
  df_test_normalized = df_test.copy()
  numerical_cols = df_train_normalized.select_dtypes(include=['float64', 'int64']).columns
  if 'ADHD_Outcome' in numerical_cols:
    numerical_cols = numerical_cols.drop('ADHD_Outcome')
  if 'Sex_F' in numerical_cols:
    numerical_cols = numerical_cols.drop('Sex_F')
  if 'target_encoded' in numerical_cols:
    numerical_cols = numerical_cols.drop('target_encoded')
  scaler = MinMaxScaler()
  scaler.fit(df_train_normalized[numerical_cols])
  df_train_normalized[numerical_cols] = scaler.transform(df_train_normalized[numerical_cols])
  df_test_normalized[numerical_cols] = scaler.transform(df_test_normalized[numerical_cols])
  df_train_normalized = pd.DataFrame(df_train_normalized, columns=df_train.columns, index=df_train.index)
  df_test_normalized = pd.DataFrame(df_test_normalized, columns=df_test.columns, index=df_test.index)
  return df_train_normalized, df_test_normalized
```
