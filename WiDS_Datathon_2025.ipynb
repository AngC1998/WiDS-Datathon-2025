```python
pre_train_df['PreInt_Demos_Fam_Child_Ethnicity'] = pre_train_df['PreInt_Demos_Fam_Child_Ethnicity'].astype(pd.Int64Dtype())
pre_train_df['PreInt_Demos_Fam_Child_Race'] = pre_train_df['PreInt_Demos_Fam_Child_Race'].astype(pd.Int64Dtype())
pre_train_df['MRI_Track_Scan_Location'] = pre_train_df['MRI_Track_Scan_Location'].astype(pd.Int64Dtype())
pre_train_df['Barratt_Barratt_P1_Edu'] = pre_train_df['Barratt_Barratt_P1_Edu'].astype(pd.Int64Dtype())
pre_train_df['Barratt_Barratt_P1_Occ'] = pre_train_df['Barratt_Barratt_P1_Occ'].astype(pd.Int64Dtype())
pre_train_df['Barratt_Barratt_P2_Edu'] = pre_train_df['Barratt_Barratt_P2_Edu'].astype(pd.Int64Dtype())
pre_train_df['Barratt_Barratt_P2_Occ'] = pre_train_df['Barratt_Barratt_P2_Occ'].astype(pd.Int64Dtype())
pre_train_df
```

```python
train_df = pd.get_dummies(pre_train_df, columns=['PreInt_Demos_Fam_Child_Ethnicity',
                                             'PreInt_Demos_Fam_Child_Race',
                                             'MRI_Track_Scan_Location',
                                             'Barratt_Barratt_P1_Edu',
                                             'Barratt_Barratt_P1_Occ',
                                             'Barratt_Barratt_P2_Edu',
                                             'Barratt_Barratt_P2_Occ'], dummy_na=False)
train_df
```

```python
'EHQ_EHQ_Total', 'ColorVision_CV_Score',
       'APQ_P_APQ_P_CP', 'APQ_P_APQ_P_ID', 'APQ_P_APQ_P_INV',
       'APQ_P_APQ_P_OPD', 'APQ_P_APQ_P_PM', 'APQ_P_APQ_P_PP',
       'SDQ_SDQ_Conduct_Problems', 'SDQ_SDQ_Difficulties_Total',
       'SDQ_SDQ_Emotional_Problems', 'SDQ_SDQ_Externalizing',
       'SDQ_SDQ_Generating_Impact', 'SDQ_SDQ_Hyperactivity',
       'SDQ_SDQ_Internalizing', 'SDQ_SDQ_Peer_Problems', 'SDQ_SDQ_Prosocial',
       'MRI_Track_Age_at_Scan'
```

```python
train_df['EHQ_EHQ_Total'].fillna(train_df['EHQ_EHQ_Total'].mean(), inplace=True)
train_df['ColorVision_CV_Score'].fillna(train_df['ColorVision_CV_Score'].median(), inplace=True)
train_df['APQ_P_APQ_P_CP'].fillna(train_df['APQ_P_APQ_P_CP'].median(), inplace=True)
train_df['APQ_P_APQ_P_ID'].fillna(train_df['APQ_P_APQ_P_ID'].median(), inplace=True)
train_df['APQ_P_APQ_P_INV'].fillna(train_df['APQ_P_APQ_P_INV'].median(), inplace=True)
train_df['APQ_P_APQ_P_OPD'].fillna(train_df['APQ_P_APQ_P_OPD'].median(), inplace=True)
train_df['APQ_P_APQ_P_PM'].fillna(train_df['APQ_P_APQ_P_PM'].median(), inplace=True)
train_df['APQ_P_APQ_P_PP'].fillna(train_df['APQ_P_APQ_P_PP'].median(), inplace=True)
train_df['SDQ_SDQ_Conduct_Problems'].fillna(train_df['SDQ_SDQ_Conduct_Problems'].median(), inplace=True)
train_df['SDQ_SDQ_Difficulties_Total'].fillna(train_df['SDQ_SDQ_Difficulties_Total'].median(), inplace=True)
train_df['SDQ_SDQ_Emotional_Problems'].fillna(train_df['SDQ_SDQ_Emotional_Problems'].median(), inplace=True)
train_df['SDQ_SDQ_Externalizing'].fillna(train_df['SDQ_SDQ_Externalizing'].median(), inplace=True)
train_df['SDQ_SDQ_Generating_Impact'].fillna(train_df['SDQ_SDQ_Generating_Impact'].median(), inplace=True)
train_df['SDQ_SDQ_Hyperactivity'].fillna(train_df['SDQ_SDQ_Hyperactivity'].median(), inplace=True)
train_df['SDQ_SDQ_Internalizing'].fillna(train_df['SDQ_SDQ_Internalizing'].median(), inplace=True)
train_df['SDQ_SDQ_Peer_Problems'].fillna(train_df['SDQ_SDQ_Peer_Problems'].median(), inplace=True)
train_df['SDQ_SDQ_Prosocial'].fillna(train_df['SDQ_SDQ_Prosocial'].median(), inplace=True)
train_df['MRI_Track_Age_at_Scan'].fillna(train_df['MRI_Track_Age_at_Scan'].median(), inplace=True)
train_df
```
