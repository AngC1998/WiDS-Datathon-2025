```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
training_solutions_df = pd.read_excel('/content/TRAINING_SOLUTIONS.xlsx')
training_solutions_df
```

```python
adhd_count = training_solutions_df.groupby('ADHD_Outcome').count()['participant_id']
adhd_count = adhd_count.to_frame()
adhd_count
```
